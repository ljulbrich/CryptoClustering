```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
import holoviews as hv
hv.extension('bokeh')
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()

scaled_market_data = scaler.fit_transform(df_market_data[['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y']])
```

```python
# Create a DataFrame with the scaled data
scaled_market_df = pd.DataFrame(scaled_market_data, columns=['price_change_percentage_24h', 'price_change_percentage_7d', 'price_change_percentage_14d', 'price_change_percentage_30d', 'price_change_percentage_60d', 'price_change_percentage_200d', 'price_change_percentage_1y'])

# Copy the crypto names from the original data
scaled_market_df['coinid'] = df_market_data.index

# Set the coinid column as index
scaled_market_df = scaled_market_df.set_index('coinid')


# Display sample data
scaled_market_df.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_means = KMeans(n_clusters=i, random_state=1)
    model = k_means.fit(scaled_market_df)
    inertia.append(model.inertia_)

```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_curve = {'k' : k, 'inertia' : inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_curve_df = pd.DataFrame(elbow_curve)
elbow_curve_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
elbow_curve_graph = elbow_curve_df.hvplot.line(
    x = 'inertia',
    y = 'k',
    title = 'Elbow curve'
)

display(elbow_curve_graph)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The best value for `k` for the maximum change in inertia for the random state 1 is most likely 4.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
model.fit(scaled_market_df)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
k_4 = model.predict(scaled_market_df)

# Print the resulting array of cluster values.
print(k_4)
```

```python
# Create a copy of the DataFrame
df_market_data_scaled = scaled_market_df.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_scaled['Predicted Cluster'] = k_4

# Display sample data
df_market_data_scaled.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
k_means_scatter = df_market_data_scaled.hvplot.scatter(
    x='price_change_percentage_24h',
    y='price_change_percentage_7d',
    by='Predicted Cluster',
    hover_cols=['coinid']
)

display(k_means_scatter)
```

---

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_3 = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
crypto_pca = pca_3.fit_transform(df_market_data_scaled)
# View the first five rows of the DataFrame. 
df_pca = pd.DataFrame(crypto_pca)
df_pca.head(5)
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
pca_ratio = pca_3.explained_variance_ratio_
print(pca_ratio)
print(pca_ratio[0] + pca_ratio[1])
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 66.2% of the total explained variance is condensed into the top principal components

```python
# Create a new DataFrame with the PCA data.
df_pca = df_pca
# Creating a DataFrame with the PCA data
df_pca = df_pca # idk why it wants me to make the df 3 times but here you go

# Copy the crypto names from the original data
df_pca['coinid'] = df_market_data.index

# Set the coinid column as index
df_pca = df_pca.set_index('coinid')

# Display sample data
df_pca.head(5)
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k
```

```python
# Create an empty list to store the inertia values
pca_inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    model = KMeans(n_clusters=i)
    model.fit(df_pca)
    pca_inertia.append(model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
pca_elbow_curve = {'k' : k, 'inertia' : pca_inertia}
# Create a DataFrame with the data to plot the Elbow curve
pca_elbow_df = pd.DataFrame(pca_elbow_curve)
pca_elbow_df.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow_graph = pca_elbow_df.hvplot.line(
    x='inertia',
    y='k',
    title='PCA elbow curve'
)

display(pca_elbow_graph)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** The best value for k in this data is 4.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** This is the same value of k that was found using the original data. This data does however have a much steeper curve, making it easier to identify.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialise the K-Means model using the best value for k
model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the PCA data
model.fit(df_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the PCA data
k_pca_4 = model.predict(df_pca)
# Print the resulting array of cluster values.
k_pca_4[:10]
```

```python
# Create a copy of the DataFrame with the PCA data
pca_k_df = df_pca
pca_k_df = pca_k_df.rename(columns={0 : 'PC1', 1 : 'PC2', 2 : 'PC3'}, errors='raise')

# Add a new column to the DataFrame with the predicted clusters
pca_k_df['predicted clusters'] = k_pca_4

# Display sample data
pca_k_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Colour the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
pca_scatter = pca_k_df.hvplot.scatter(
    x='PC1',
    y='PC2',
    by='predicted clusters',
    hover_cols='coinid'
)

display(pca_scatter)
```
